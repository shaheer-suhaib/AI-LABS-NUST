```python
from tensorflow import layers,models



# model
model = models.sequential()
models.add(layers.conv2d(32, (3, 3), activation='relu', input_shape=(64, 64, 3)))
models.add(layers.maxpooling2d((2, 2)))

models.add(layers.conv2d(64, (3, 3), activation='relu'))  #1
model.add(layers.maxpooling2d((2, 2)))

models.add(layers.conv2d(128, (3, 3), activation='relu'))
models.add(layers.maxpooling2d((2, 2)))

models.add(layers.conv2d(128, (3, 3), activation='relu'))
models.add(layers.maxpooling2d((2, 2)))
models.add(layers.flatten())

models.add(layers.dense(1024, activation='relu'))
# binary class
models.add(layers.dense(1, activation='sigmoid'))

model.summary()



#


```

```python
# now compiling
#  Mulit calssification
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```
